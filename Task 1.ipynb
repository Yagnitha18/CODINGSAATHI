# Load the data

```python
# Iris Flower Classification ML Project
# Import Packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels'] 
# Load the data
df = pd.read_csv('iris.data', names=columns)
df.head()
```

# Analyze and visualize the dataset

```python
# Some basic statistical analysis about the data
df.describe()
```

```python
# Assuming you already have the 'df' DataFrame loaded with the data

# Pairplot - Visualization of pairwise relationships in the dataset
sns.pairplot(df, hue='Class_labels', markers='o')
plt.title('Pairplot of Iris Dataset')
plt.show()

# Box Plot - Visualization of the distribution of each feature by class label
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Class_labels', y='Petal length')
plt.title('Box Plot - Petal length by Class Labels')
plt.show()

# Scatter Plot - Visualization of Sepal length vs Sepal width
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Sepal length', y='Sepal width', hue='Class_labels', style='Class_labels')
plt.title('Scatter Plot - Sepal length vs Sepal width')
plt.show()

# Histograms - Visualization of feature distributions
df.hist(figsize=(10, 6))
plt.suptitle('Histograms of Iris Dataset Features', y=1.02)
plt.tight_layout()
plt.show()

# Heatmap - Correlation matrix of features
correlation_matrix = df.drop('Class_labels', axis=1).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Separate features and target  
data = df.values
X = data[:,0:4]
Y = data[:,4]
```

```python
# Calculate average of each features for all classes
Y_Data = np.array([np.average(X[:, i][Y==j].astype('float32')) for i in range (X.shape[1])
 for j in (np.unique(Y))])
Y_Data_reshaped = Y_Data.reshape(4, 3)
Y_Data_reshaped = np.swapaxes(Y_Data_reshaped, 0, 1)
X_axis = np.arange(len(columns)-1)
width = 0.25
```

```python
# Plot the average
plt.bar(X_axis, Y_Data_reshaped[0], width, label = 'Setosa')
plt.bar(X_axis+width, Y_Data_reshaped[1], width, label = 'Versicolour')
plt.bar(X_axis+width*2, Y_Data_reshaped[2], width, label = 'Virginica')
plt.xticks(X_axis, columns[:4])
plt.xlabel("Features")
plt.ylabel("Value in cm.")
plt.legend(bbox_to_anchor=(1.3,1))
plt.show()
```

# Model Training

```python
# Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Support vector machine algorithm
from sklearn.svm import SVC
svn = SVC()
svn.fit(X_train, y_train)
```

# Model Evaluation

```python
# Predict from the test dataset
predictions = svn.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
# A detailed classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
